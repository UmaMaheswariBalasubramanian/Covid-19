# Top 10 States with highest total cases

```python
high_cases = df_covid.sort_values(by = 'Total Cases',ascending=False).head(10)
xpos = np.arange(len(high_cases['State/UTs']))
fig = plt.subplots(figsize=(15,6))
plt.xticks(xpos,high_cases['State/UTs'])
plt.bar(high_cases['State/UTs'],high_cases['Total Cases'],color = 'red',width=0.5)
plt.xlabel('Name of the states')
plt.ylabel('Total Cases')
plt.title("Top 10 states by Total cases")
plt.show()
```

# Top 5 States with highest total cases

```python
least_cases = df_covid.sort_values(by = 'Total Cases',ascending=True).head()
ypos = np.arange(len(least_cases['State/UTs']))
fig = plt.subplots(figsize=(20,6))
plt.xticks(ypos,least_cases['State/UTs'])
plt.bar(least_cases['State/UTs'],least_cases['Total Cases'],color = 'blue',width=0.5)
plt.xlabel('Name of the states')
plt.ylabel('Total Cases')
plt.title("Top 5 states by Total cases")
plt.show()
```

# Top 10 states having highest number of deaths

```python
high_deaths = df_covid.sort_values(by = 'Deaths',ascending=False).head(10)
xpos = np.arange(len(high_cases['State/UTs']))
plt.figure(figsize=(15,8))
plt.scatter(x = high_deaths['State/UTs'],y = high_deaths['Deaths'],c= 'black')
plt.plot(high_deaths['State/UTs'],high_deaths['Deaths'],c='r')
plt.xlabel('Name of the states')
plt.ylabel('Deaths')
plt.title("Top 10 states with highest no. of Deaths")
plt.show()
plt.grid()
```
