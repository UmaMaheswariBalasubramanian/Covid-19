Now performing Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()  #Creating instance
df_covid['State_No'] = labelencoder.fit_transform(df_covid['State/UTs']) # New numerical column
df_covid
```

# Now performing One-Hot Encoding

```python
df_covid = pd.get_dummies(df_covid['State/UTs'])
df_covid
```

# Performing Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
df2=df_covid
```

```python
df2.columns
```

```python
num_cols = ['Total Cases', 'Active', 'Discharged', 'Deaths'] #numerical features(dtypes:int)
for i in num_cols: #Apply Standardization on numerical features
    scale = StandardScaler().fit(df2[[i]])  #fit on training data column
    df2[[i]] = scale.transform(df2[[i]])   #transform the training data column
```
